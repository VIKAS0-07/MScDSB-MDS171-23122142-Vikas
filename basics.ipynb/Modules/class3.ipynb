```python
class sportMart:
    def __init__(self):
        self.inventory = {}
        self.orders = {}

    def createOrder(self,OrderID,ItemName,Quantity,Price):
        temp={
            "orderid" : OrderID,
            "itemName" : ItemName,
            "quantity" : Quantity,
            "price" : Price
        }
        self.orders[OrderID]=temp

    def createInventoryItem(self,ProductId,ProductName,Quantity,Price):
        temp={
            "Product Id" : ProductId,
            "Product_Name" : ProductName,
            "quantity" : Quantity,
            "price" : Price
        }
        self.inventory[ProductId] = temp

    def printOrders(self):
        print(self.orders)

    def PrintInventory(self):
        print(self.inventory)

    def Update_Orders(self,OrderID,ItemName,Quantity,Price):
        temp={
            "orderid" : OrderID,
            "itemName" : ItemName,
            "quantity" : Quantity,
            "price" : Price
        }
        #Create a menudriven program # Create new orders and update inventory #Export the final data to the fill
        self.orders[OrderID]=temp
        print(self.orders)

    def write_file(self):
        cust=open("customer_order.csv","w+")
        for key in self.inventory:
            cust.write(self.inventory[key]["productid"]+","+self.inventory[key]["itemName"]+","+str(self.inventory[key]["quantity"])+","+str(self.inventory[key]["price"])+"\n")
        cust.close()

trinity = sportMart()
orders = open("customer_order.csv","r+")
orders_header=orders.readline()
orders_orders=orders.readlines()
for item in orders_orders:
    temp=item.strip().split(',')
    trinity.createOrder(temp[0],temp[1],temp[2],temp[3])
trinity.printOrders()

while True:
    choice=int(input("Enter your choice:"))
    if choice==1:
        OrderID=input("Enter the Product ID:")
        ItemName=input("Enter the product name:")
        Quantity=input("Enter the quantity:")
        Price=input("Enter the price of the product:")
        trinity.Update_Orders(OrderID,ItemName,Quantity,Price)
    elif choice==2:
        trinity.write_file()
        exit()
```
