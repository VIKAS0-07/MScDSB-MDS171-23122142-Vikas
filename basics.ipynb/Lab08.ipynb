Write a Python program to implement a Stack class that has the following functions

Push Item

Pop Item

Print the Items in the stack

Size of Stack

The top item in the stack

Check if the stack is empty

```python
class stack_class:
    def __init__(self):
        self.stack=[]
        
    def push_item(self):
        self.stack.append("APPLE")
        self.stack.append("ORANGE")  
        self.stack.append("PINEAPPLE")
        self.stack.append("GRAPES")
        
        print("Initial stack:")
        print(self.stack)

    def pop_item(self):
        print("Elements that are popped")

        print(self.stack.pop())
        print(self.stack.pop())
        
    def disp_item(self):
        print("Elements in stack:")
        print(self.stack)

    def size_stack(self):
        print("The size of the stack:")
        print(len(self.stack))

    def top_item(self):
        print("The top item of the stack is:")
        print(self.stack[-1])

    def stack_search(self):
        if len(self.stack)==0:
            print("The stack is empty")
        else:
            print("The elements in stack are not empty")
    
element=stack_class()

while True:
    print("....STACK CLASS.....")
    print("1.Print the elements that are added:")
    print("2.Print the elements that are popped:")
    print("3.Print the elements in the stack:")
    print("4.Print the size of the stack:")
    print("5.Print the top item in stack:")
    print("6.Check if the stack is empty:")
    print("7.Exit")

    choice=int(input("Enter your choice:"))

    if choice==1:
        element.push_item()

    elif choice==2:
        element.pop_item()

    elif choice==3:
        element.disp_item()

    elif choice==4:
        element.size_stack()

    elif choice==5:
        element.top_item()

    elif choice==6:
        element.stack_search()

    elif choice==7:
        exit()
    else:
        print("Invalid choice")


    


    
```
