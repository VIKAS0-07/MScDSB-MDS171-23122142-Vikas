Create a class pandas name that will help read a csv file and print the top/bottom n rows in the file

Write a python program with the following specifications .

There will be a function named create dataset which will take a dictionary,filename,number of rows as arguments.

The dictionary is having the following structure {
    "col name" : [domain and values]
    "    "
}

The function will select random values for each of the domains and create a row for you.

The number of rows that has to be created is a parameter.

Save the file with the required number of rows.

Load this dataset using pandas and display basic information{which includes shape,size etc} 

```python
import pandas as pd
class trinity:

    def __init__(self):
        self.dict={
            "Orderid" : ["6321","1125","2314"],
            "ItemName" : ["Munch","Twix","KitKat"],
            "Price" : [20,35,40],
            "Quantity" : [5,10,20],
            "ShelfNumber" : [24,31,26],
            }
    def create_datset(self):
        file=open("trinity.csv","w+")
        file.write("Orderid,ItemName,Price,Quantity,Shelf Number" "\n".title())
        import random
        for i in range(0,50,1):
            a=self.dict["Orderid"][random.randint(0,len(self.dict["Orderid"])-1)]
            b=self.dict["ItemName"][random.randint(0,len(self.dict["ItemName"])-1)]
            c=self.dict["Price"][random.randint(0,len(self.dict["Price"])-1)]
            d=self.dict["Quantity"][random.randint(0,len(self.dict["Quantity"])-1)]
            e=self.dict["ShelfNumber"][random.randint(0,len(self.dict["ShelfNumber"])-1)]
            file.write(str(a)+","+str(b)+","+str(c)+","+str(d)+","+str(e)+"\n")
        file=open("trinity.csv","a+")
        file.close()




order=trinity()
order.create_datset()
rc=pd.read_csv("trinity.csv")
rc.head()  

```
